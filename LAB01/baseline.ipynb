```python
import pandas as pd
from pathlib import Path
import random
from ultralytics import YOLO


def generate_predictions():
  # Load your YOLOv8 classification model
  model = YOLO('/content/drive/MyDrive/40epochs_128batch/runs/classify/train/weights/best.pt')

  # Define the folder containing images
  image_folder = '/content/to_predict'

  # Initialize a list to store results
  results = []
  i = 0
  # Iterate over all  image files in the folder
  for image_file in Path(image_folder).glob('*.*'):
    # Get the image name
    image_name = image_file.name

    # Check if the file has an .md extension
    if image_file.suffix == '.md' or image_file.suffix == '.HEIC':
        predicted_class = random.randint(0, 9)
    else:
        # Make a prediction on the image
        prediction = model(image_file)

        # Assuming the prediction returns a single class with the highest confidence
        predicted_class = prediction[0].probs.top1

    # Append the result to the list in the desired format
    results.append(f"{image_name},{predicted_class}")

    # Save the results to a CSV file
    with open('predictions1.csv', 'w') as file:
        for result in results:
            file.write(result + '\n')

    print(i)
    i += 1
```

```python
generate_predictions()
```
